2. Write a Python Program to Find HCF?

```python
print("enter two numbers:")
a,b=int(input()),int(input())
c=max(a,b)
l=[]
for i in range(1,c+1):
  if a%i==0 and b%i==0:
    l.append(i)
print(f"the HCF is {max(l)}")
```

3. Write a Python Program to Convert Decimal to Binary, Octal and Hexadecimal?

```python
a=int(input("The decimal number is:"))
print(f"In binary the number is {bin(a)}")
print(f"In octal the number is {oct(a)}")
print(f"In hexadecimal the number is {hex(a)}")
```
