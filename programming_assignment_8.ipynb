1. Write a Python Program to Add Two Matrices?

```python
import numpy as np  
#first matrix  
Rows = int(input("Give the number of rows:"))  
Columns = int(input("Give the number of columns:"))  
print("Please write the elements of the matrix in a single line and separated by a space: ")    
elements = list(map(int, input().split()))   
matrix1 = np.array(elements).reshape(Rows, Columns)  
#second matrix
Rows = int(input("Give the number of rows:"))  
Columns = int(input("Give the number of columns:"))  
print("Please write the elements of the matrix in a single line and separated by a space: ")    
elements = list(map(int, input().split()))   
matrix2 = np.array(elements).reshape(Rows, Columns) 
sum= matrix1+matrix2
print(f"the sum is \n {sum}")
```

2. Write a Python Program to Multiply Two Matrices?


```python
import numpy as np  
#first matrix  
Rows = int(input("Give the number of rows:"))
Columns = int(input("Give the number of columns:"))  
print("Please write the elements of the matrix in a single line and separated by a space: ")    
elements = list(map(int, input().split()))   
matrix1 = np.array(elements).reshape(Rows, Columns)  
#second matrix
Rows = int(input("Give the number of rows:"))  
Columns = int(input("Give the number of columns:"))  
print("Please write the elements of the matrix in a single line and separated by a space: ")    
elements = list(map(int, input().split()))   
matrix2 = np.array(elements).reshape(Rows, Columns) 
pro=np.dot(matrix1,matrix2)
print(f"the product is \n {pro}")
```

3. Write a Python Program to Transpose a Matrix?


```python
import numpy as np   
Rows = int(input("Give the number of rows:"))
Columns = int(input("Give the number of columns:"))  
print("Please write the elements of the matrix in a single line and separated by a space: ")    
elements = list(map(int, input().split()))   
matrix = np.array(elements).reshape(Rows, Columns)
T=matrix.transpose()
print(T)
```

5. Write a Python Program to Remove Punctuation From a String?

```python
punctuations = '''!@#$%^&*()_+-={}:"|<>?/.,\';][*-+]'''
a=input("Enter a string: ")
l=[]
for i in a:
   if i not in punctuations:
       l.append(i)
str2=''.join(l)
print(str2)
```
