9. Write a Python program to Remove empty List from List?

```python
l= ['sam','rat',6 , [],7,8, [], 9]
print(f"The original list is : {str(l)}")
while [] in l :
	l.remove([])
print(f"List after empty list removal : {str(l)}")
```

10. Write a Python program to Cloning or Copying a list?

```python
l=[]
n=int(input("Enter the number of elements:"))
for i in range(n):
  l.append(input())
l1=l.copy()
print(f"First list is  :{l1}")
print(f"After copying the copied list :{l1}")
```

11. Write a Python program to Count occurrences of an element in a list?

```python
l=[]
n=int(input("Enter the number of elements:"))
for i in range(n):
  l.append(input())
n=input("enter the element:")
a=l.count(n)
print(f"The occurrence number of the elementis: {a}")
```
