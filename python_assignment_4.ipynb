1. Write a Python Program to Find the Factorial of a Number?

```python
a=int(input('Enter the number:'))
if a==0:
  print("Factorial:1")
else:
  pro=1
  for i in range(1,a+1):
    pro*=i
  print(f"factorial: {pro}")
```

3. Write a Python Program to Print the Fibonacci sequence?

```python
a=int(input('Enter the number of elements you want to show for fibonacci sequence:'))
c=0
b=1
l=[]
for i in range(a):
  l.append(c)
  c,b=b,c+b

print(l)

```

4. Write a Python Program to Check Armstrong Number?

```python
#An Armstrong number of three digits is an integer such that the sum of the cubes of its digits is equal to the number itself.
a = int(input("Enter a number: "))
sum=0
b=a
while b > 0:
   digit=b%10
   sum+=digit**3
   b//=10
if a==sum:
   print("It is an Armstrong number")
else:
   print("It is not an Armstrong number")
```

5. Write a Python Program to Find Armstrong Number in an Interval?

```python
print("Enter the interval")
c,d=map(int,input().split())
for a in range(c,d+1):
    sum=0
    b=a
    while b > 0:
      digit=b%10
      sum+=digit**3
      b//=10
    if a==sum:
      print(a)

```

6. Write a Python Program to Find the Sum of Natural Numbers?

```python
a = int(input("Enter the natural number upto which you need the sum: "))
sum=0
for i in range(a+1):
  sum=sum+i
print(sum)
```
