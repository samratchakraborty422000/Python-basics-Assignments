```python
a,b=map(float,input("enter base and height:").split())
area=1/2*a*b
print(f"area= {area}")
```

4. Write a Python program to swap two variables?

```python
a,b= input(),input()
print(f"before swap a={a},b={b}")
a,b=b,a
print(f"after swap a={a},b={b}")
```
